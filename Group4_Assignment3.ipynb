```python
#Ex1. BMI Function
def calculate_bmi(weight_lbs, height_ft, height_in):
    weight_kg = weight_lbs * 0.453592
    height_m = (height_ft * 12 + height_in) * 0.0254
    bmi = weight_kg / (height_m ** 2)

    if bmi < 18.5:
        category = "Underweight"
    elif bmi < 25:
        category = "Normal"
    elif bmi < 30:
        category = "Overweight"
    else:
        category = "Obesity"

    return round(bmi, 2), category

while True:
    weight = float(input("\nEnter weight in pounds (0 to quit): "))
    if weight == 0:
        break

    height_ft = float(input("Enter height in feet: "))
    height_in = float(input("Enter additional inches: "))

    bmi_value, bmi_category = calculate_bmi(weight, height_ft, height_in)
    print(f"Your BMI is {bmi_value} ({bmi_category}).")
```

```python
#Ex2. Purchase Analytics
customers = [
    {'CID': 1001, 'age': 30, 'email': 'a@gmail.com', 'phone': '123-111-0001',
     'address': '123 X Street, Athens, GA 30601',
     'last 5 purchases': {'A12': 45.69, 'B478': 580.20, 'E4780': 91.00, 'C780': 140.25, 'D41': 230.14}},
    {'CID': 1002, 'age': 45, 'email': 'b@gmail.com', 'phone': '123-111-0002',
     'address': '200 Zys Street, Atlanta, GA 30301',
     'last 5 purchases': {'B235': 80.12, 'A12': 50.91, 'C780': 175.25, 'E124': 132.25, 'D20': 49.87}},
    {'CID': 1003, 'age': 29, 'email': 'c@gmail.com', 'phone': '123-111-0003',
     'address': '10 Oak Ave, Savannah, GA 31401',
     'last 5 purchases': {'A12': 55.00, 'B222': 99.00, 'C333': 10.00, 'D444': 60.00, 'E555': 40.00}},
    {'CID': 1004, 'age': 37, 'email': 'd@gmail.com', 'phone': '123-111-0004',
     'address': '77 Pine Road, Marietta, GA 30060',
     'last 5 purchases': {'A10': 90.00, 'B20': 60.00, 'C30': 75.00, 'D40': 30.00, 'E50': 25.00}},
    {'CID': 1005, 'age': 52, 'email': 'e@gmail.com', 'phone': '123-111-0005',
     'address': '500 Lake Dr, Macon, GA 31201',
     'last 5 purchases': {'A12': 25.50, 'B235': 90.00, 'C800': 60.00, 'D40': 120.00, 'E124': 45.00}},
    {'CID': 1006, 'age': 41, 'email': 'f@gmail.com', 'phone': '123-111-0006',
     'address': '9 Elm St, Columbus, GA 31901',
     'last 5 purchases': {'A12': 70.00, 'B300': 40.00, 'C400': 35.00, 'D200': 25.00, 'E124': 60.00}},
    {'CID': 1007, 'age': 50, 'email': 'g@gmail.com', 'phone': '123-111-0007',
     'address': '12 River Rd, Augusta, GA 30901',
     'last 5 purchases': {'A10': 45.00, 'B20': 85.00, 'C30': 100.00, 'D40': 35.00, 'E50': 90.00}},
    {'CID': 1008, 'age': 33, 'email': 'h@gmail.com', 'phone': '123-111-0008',
     'address': '88 Harbor Ln, Tybee Island, GA 31328',
     'last 5 purchases': {'A12': 50.00, 'B22': 60.00, 'C33': 70.00, 'D44': 80.00, 'E55': 90.00}},
    {'CID': 1009, 'age': 26, 'email': 'i@gmail.com', 'phone': '123-111-0009',
     'address': '400 College St, Athens, GA 30602',
     'last 5 purchases': {'A10': 65.00, 'B20': 75.00, 'C30': 85.00, 'D40': 95.00, 'E50': 105.00}},
    {'CID': 1010, 'age': 39, 'email': 'j@gmail.com', 'phone': '123-111-0010',
     'address': '210 Main St, Albany, GA 31701',
     'last 5 purchases': {'A12': 30.00, 'B222': 45.00, 'C333': 60.00, 'D444': 75.00, 'E555': 90.00}}
]

# 1. Customer Purchase Summary
def customer_purchase_summary(customers):
    summary = {}
    for customer in customers:
        total = sum(customer['last 5 purchases'].values())
        summary[customer['CID']] = round(total, 2)
    return summary

summary = customer_purchase_summary(customers)
print("1) Customer Purchase Summary (CID : total):")
for cid, total in summary.items():
    print(f" CID {cid}: ${total}")

# 2. Find Customers with Max and Min Totals
def find_max_min(purchase_summary):
    max_cid = max(purchase_summary, key=purchase_summary.get)
    min_cid = min(purchase_summary, key=purchase_summary.get)
    return (max_cid, purchase_summary[max_cid]), (min_cid, purchase_summary[min_cid])

max_pair, min_pair = find_max_min(summary)
print("\n2) Max and Min Totals:")
print(f" Highest: CID {max_pair[0]} with ${max_pair[1]}")
print(f" Lowest:  CID {min_pair[0]} with ${min_pair[1]}")

# 3. Filter Customers by Spending Threshold
def filter_customers_by_spending(purchase_summary, x):
    return {cid: total for cid, total in purchase_summary.items() if total < x}

print("\n3) Filter Customers by Spending Threshold:")
for x in [100, 300, 500, 1000]:
    filtered = filter_customers_by_spending(summary, x)
    print(f" Totals < ${x}:")
    if filtered:
        for cid, total in filtered.items():
            print(f"  CID {cid}: ${total}")
    else:
        print("  None")

# 4. Count Customers Above a Spending Threshold
def count_customers_above_threshold(purchase_summary, x):
    count = 0
    for total in purchase_summary.values():
        if total > x:
            count += 1
    return count

print("\n4) Count Customers Above a Spending Threshold:")
for x in [100, 300, 500, 1000]:
    count = count_customers_above_threshold(summary, x)
    print(f" Customers with totals > ${x}: {count}")

# 5. Analyze Purchases by Item Initial
def analyze_purchases_by_letter(purchases):
    totals = {'A': 0, 'B': 0, 'C': 0, 'D': 0, 'E': 0}
    for code, price in purchases.items():
        letter = code[0].upper()
        if letter in totals:
            totals[letter] += price
    highest_letter = max(totals, key=totals.get)
    lowest_letter = min(totals, key=totals.get)
    return totals, (highest_letter, round(totals[highest_letter], 2)), (lowest_letter, round(totals[lowest_letter], 2))

print("\n5) Analyze Purchases by Item Initial for Each Customer:")
for cust in customers:
    totals, highest, lowest = analyze_purchases_by_letter(cust['last 5 purchases'])
    print(f"\n CID {cust['CID']} totals: { {k: round(v,2) for k,v in totals.items()} }")
    print(f"  Highest initial: {highest[0]} -> ${highest[1]}")
    print(f"  Lowest initial:  {lowest[0]} -> ${lowest[1]}")

# 6. Sum Purchases by Item Initial for a Specific Customer
def sum_purchases_by_letter(customers, letter, cid):
    letter = letter.upper()
    for customer in customers:
        if customer['CID'] == cid:
            total = 0
            for code, price in customer['last 5 purchases'].items():
                if code.upper().startswith(letter):
                    total += price
            return round(total, 2)
    return 0.0

print("\n6) Sum Purchases by Item Initial for a Specific Customer:")
print(f" CID 1001, letter 'A' -> ${sum_purchases_by_letter(customers, 'A', 1001)}")
print(f" CID 1002, letter 'C' -> ${sum_purchases_by_letter(customers, 'C', 1002)}")
print(f" CID 1009, letter 'E' -> ${sum_purchases_by_letter(customers, 'E', 1009)}")

# 7. Statistical Analysis by Item Initial
def stats_by_item_initial(customers, letter):
    letter = letter.upper()
    prices = []
    for customer in customers:
        for code, price in customer['last 5 purchases'].items():
            if code.upper().startswith(letter):
                prices.append(price)
    if not prices:
        return 0.0, 0.0, 0.0
    prices.sort()
    total = sum(prices)
    avg = total / len(prices)
    n = len(prices)
    if n % 2 == 1:
        median = prices[n // 2]
    else:
        median = (prices[n // 2 - 1] + prices[n // 2]) / 2
    return round(total, 2), round(avg, 2), round(median, 2)

print("\n7) Statistical Analysis by Item Initial Across All Customers:")
for letter in ['A', 'B', 'C', 'D', 'E']:
    total, avg, median = stats_by_item_initial(customers, letter)
    print(f" {letter}: Total=${total}, Avg=${avg}, Median=${median}")
```
