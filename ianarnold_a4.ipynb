```python
#8-7. Album
def make_album(artist, title, tracks=None):
    album = {"artist": artist, "title": title}
    if tracks:
        album["tracks"] = tracks
    return album

print(make_album("Queen", "Jazz"))
print(make_album("Adele", "30"))
print(make_album("Kanye West", "Graduation", 13))
```

```python
#8-10 (1). Sending Messages
magicians = ["houdini", "blaine", "angel"]
sent_messages = []

def show_magicians(names):
    for n in names:
        print(n.title())

def send_messages(messages, sent_messages):
    """Print each message (magician name) and move it to sent_messages."""
    while messages:
        message = messages.pop()             
        print(f"Sending message: {message.title()}")
        sent_messages.append(message)

show_magicians(magicians[:]) 
send_messages(magicians, sent_messages)
```

```python
#8-11. Unchanged Magicians
def make_great(names):
    return [f"The Great {name}." for name in names]

def show_magicians(names):
    for name in names:
        print(name)

magicians = ["houdini", "blaine", "angel"]
great_magicians = make_great(magicians[:])

show_magicians(magicians)
show_magicians(great_magicians)
```

```python
#9-1. Restaurant
class Restaurant:
    def __init__(self, name, cuisine):
        self.restaurant_name = name
        self.cuisine_type = cuisine

    def describe_restaurant(self):
        print(f"{self.restaurant_name} serves {self.cuisine_type}.")

    def open_restaurant(self):
        print(f"{self.restaurant_name} is open!")

restaurant = Restaurant("Popeyes", "Fast Food")
print(restaurant.restaurant_name)
print(restaurant.cuisine_type)
restaurant.describe_restaurant()
restaurant.open_restaurant()
```

```python
#9-3. Users
class User:
    def __init__(self, first, last, age, city, email, username):
        self.first_name = first
        self.last_name = last
        self.age = age
        self.city = city
        self.email = email
        self.username = username

    def describe_user(self):
        print(f"{self.first_name} {self.last_name}, {self.age}, from {self.city}, email: {self.email}, username: {self.username}")

    def greet_user(self):
        print(f"Welcome back, {self.first_name}!")

user1 = User("Ian", "Arnold", 23, "Athens", "ian@example.com", "ian22")
user2 = User("John", "Mauro", 21, "Savannah", "john@example.com", "john21")
user3 = User("Chris", "Kazunas", 21, "Medford", "chris@example.com", "chris21")

user1.describe_user()
user1.greet_user()

user2.describe_user()
user2.greet_user()

user3.describe_user()
user3.greet_user()
```

```python
#9-4. Number Served
class Restaurant:
    def __init__(self, restaurant_name, cuisine_type):
        self.restaurant_name = restaurant_name
        self.cuisine_type = cuisine_type
        self.number_served = 0

    def describe_restaurant(self):
        print(f"{self.restaurant_name} serves {self.cuisine_type} cuisine.")

    def open_restaurant(self):
        print(f"{self.restaurant_name} is now open!")

    def set_number_served(self, number):
        self.number_served = number

    def increment_number_served(self, additional_customers):
        self.number_served += additional_customers


restaurant = Restaurant("Tasty Bites", "Italian")
print(f"Customers served: {restaurant.number_served}")

restaurant.number_served = 15
print(f"Customers served: {restaurant.number_served}")

restaurant.set_number_served(30)
print(f"Customers served: {restaurant.number_served}")

restaurant.increment_number_served(25)
print(f"Customers served: {restaurant.number_served}")
```

```python
#9-5. Login Attempts
class User:
    def __init__(self, first_name, last_name, age, location):
        self.first_name = first_name
        self.last_name = last_name
        self.age = age
        self.location = location
        self.login_attempts = 0

    def describe_user(self):
        print(f"User: {self.first_name} {self.last_name}, Age: {self.age}, Location: {self.location}")

    def greet_user(self):
        print(f"Welcome back, {self.first_name}!")

    def increment_login_attempts(self):
        self.login_attempts += 1

    def reset_login_attempts(self):
        self.login_attempts = 0


user = User("Ian", "Arnold", 23, "Georgia")
user.increment_login_attempts()
user.increment_login_attempts()
user.increment_login_attempts()
print(f"Login attempts: {user.login_attempts}")

user.reset_login_attempts()
print(f"Login attempts after reset: {user.login_attempts}")
```

```python
#9-6. Ice Cream Stand
class IceCreamStand(Restaurant):
    def __init__(self, restaurant_name, cuisine_type):
        super().__init__(restaurant_name, cuisine_type)
        self.flavors = ["vanilla", "chocolate", "strawberry", "mint"]

    def display_flavors(self):
        print("Available flavors:")
        for flavor in self.flavors:
            print(f"- {flavor}")


ice_cream_stand = IceCreamStand("Dairy Queen", "Dessert")
ice_cream_stand.describe_restaurant()
ice_cream_stand.display_flavors()
```

```python
#9-7. Admin
class Admin(User):
    def __init__(self, first_name, last_name, age, location):
        super().__init__(first_name, last_name, age, location)
        self.privileges = ["can add post", "can delete post", "can ban user"]

    def show_privileges(self):
        print("Admin privileges:")
        for privilege in self.privileges:
            print(f"- {privilege}")


admin_user = Admin("Sarah", "Jones", 30, "Arizona")
admin_user.show_privileges()
```

```python
#9-8. Privileges
class Privileges:
    def __init__(self, privileges=None):
        if privileges is None:
            privileges = ["can add post", "can delete post", "can ban user"]
        self.privileges = privileges

    def show_privileges(self):
        print("Privileges:")
        for privilege in self.privileges:
            print(f"- {privilege}")


class Admin(User):
    def __init__(self, first_name, last_name, age, location):
        super().__init__(first_name, last_name, age, location)
        self.privileges = Privileges()


admin_user = Admin("Sarah", "Jones", 30, "Arizona")
admin_user.privileges.show_privileges()
```

```python
#9-14. Dice
from random import randint

class Die:
    def __init__(self, sides=6):
        self.sides = sides

    def roll_die(self):
        print(randint(1, self.sides))
        
die_6 = Die()
for i in range(10):
    die_6.roll_die()

die_10 = Die(10)
for i in range(10):
    die_10.roll_die()

die_20 = Die(20)
for i in range(10):
    die_20.roll_die()
```
