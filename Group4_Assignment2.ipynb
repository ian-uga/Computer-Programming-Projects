```python
#5 Count customers with any purchase > 100 and list their CIDs
cids_over_100 = []
for info in customers.values():
    purchases = info["last 5 purchases"]
    has_big = False
    for cost in purchases.values():
        if cost > 100:
            has_big = True
            break  # early exit
    if has_big:
        cids_over_100.append(info["CID"])
print("Customers with at least one purchase > 100:", len(cids_over_100))
print("   CIDs:", cids_over_100, "\n")
```

```python
#6 Total for codes starting with 'B'
total_B = 0.0
for info in customers.values():
    for code, cost in info["last 5 purchases"].items():
        if len(code) > 0 and code[0] == "B":
            total_B = total_B + cost
print("Total spent on items starting with 'B':", total_B, "\n")
```

```python
#7 Totals by initial and highest/lowest letters
letters = ["A", "B", "C", "D", "E"]
totals_by_letter = {}
for L in letters:
    totals_by_letter[L] = 0.0

for info in customers.values():
    for code, cost in info["last 5 purchases"].items():
        first = code[0]
        if first in totals_by_letter:
            totals_by_letter[first] = totals_by_letter[first] + cost

print("Totals by item initial:")
for L in letters:
    print("   " + L + ":", totals_by_letter[L])

max_letter_value = None
min_letter_value = None
max_letters = []
min_letters = []
for L in letters:
    v = totals_by_letter[L]
    if max_letter_value is None or v > max_letter_value:
        max_letter_value = v
        max_letters = [L]
    elif v == max_letter_value:
        max_letters.append(L)

    if min_letter_value is None or v < min_letter_value:
        min_letter_value = v
        min_letters = [L]
    elif v == min_letter_value:
        min_letters.append(L)

print("   Highest:", ", ".join(max_letters), "(" + str(max_letter_value) + ")")
print("   Lowest:", ", ".join(min_letters), "(" + str(min_letter_value) + ")\n")
```

```python
#8 Sum and average, median for 'A' items
A_costs = []
for info in customers.values():
    for code, cost in info["last 5 purchases"].items():
        if code[0] == "A":
            A_costs.append(cost)

sum_A = 0.0
count_A = 0
for v in A_costs:
    sum_A = sum_A + v
    count_A = count_A + 1

if count_A > 0:
    avg_A = sum_A / count_A
else:
    avg_A = 0.0

temp = []
for v in A_costs:
    temp.append(v)

ordered = []
while True:
    n_temp = 0
    for _ in temp:
        n_temp = n_temp + 1
    if n_temp == 0:
        break

    smallest = temp[0]
    for v in temp:
        if v < smallest:
            smallest = v

    new_temp = []
    removed = False
    for v in temp:
        if (not removed) and v == smallest:
            removed = True
        else:
            new_temp.append(v)
    temp = new_temp
    ordered.append(smallest)

n = 0
for _ in ordered:
    n = n + 1

if n == 0:
    median_A = 0.0
else:
    if n % 2 == 1:
        mid_index = n // 2
        i = 0
        for v in ordered:
            if i == mid_index:
                median_A = v
            i = i + 1
    else:
        left_index = n // 2 - 1
        right_index = n // 2
        i = 0
        left_value = None
        right_value = None
        for v in ordered:
            if i == left_index:
                left_value = v
            if i == right_index:
                right_value = v
            i = i + 1
        median_A = (left_value + right_value) / 2.0

print("Items starting with 'A':")
print("   Total =", sum_A)
print("   Average =", avg_A)
print("   Median =", median_A, "\n")
```
